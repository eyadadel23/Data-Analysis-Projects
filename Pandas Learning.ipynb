ABOVE IS A SIMPLE WAY TO MODIFY ALL ROWS IN COLUMN, BUT WHAT ABOUT IF WE WANTED TO DO MORE ADVANCED AND COMPLICATED STUFF LIKE MATHMATICAL OPERATION OR ANY ADVANCED STUFF AT ONCE, WE CAN USE THE NEXT METHODES:

FIRST:(apply method): apply is used to applying a function on our values, it can work with a series or a dataframe, but the behaviour is different with each of these two objects

```python
# first, apply with a pandas series:

"""when using "apply()" with a seies, it applies a fuction on every value in that series, so we can
create any fuction we want and then use it within apply to apply it in every value in the series"""


df["original_title"].apply(len)
#the code above will apply "len" fuction on every row of "original_title" column, and print the result
```

```python
df.head(2)
```

```python
# for simple tasks or functions we can use "lambda" function with apply,exa:

df["release_year"].apply(lambda x: x+1)
# the code above will add a year to every year in "release_year" column
```

```python
# second, apply with dataframes:

"""# when using apply with a dataframe, the fuction will not run in every value in the data frame, 
insted it will run in every series(column) in that dataframe,not every row,exa:"""

df.apply(len)
# the code above will give us the "len" of every column in our datframe,not the "len" of every row of
#... every column
```

SO THE CONCLUSION IS THAT WE BETTER USE "apply()" WHEN WE WANT TO PERFORM SPECIFIC TASK OR FUNCTION ON SPECIFIC SERIES OR SPECIFIC NUMBER OF DATAFRAME COLUMNS, NOT THE HOLE DATFRAME AT ONCE

##################################################################################################

SECOND:(applymap method): applymap is used to applying a function on every indivisual element in our dataframe, it works only with a dataframe, On the contrary of "apply" methood above,we can use this one to apply a function on every row in every column all at once, NOTE: all dataframe values must be the same data type to use "applymap" correctally

```python
# lets first create pandas dataframes to simplfy things:
students_info = ({"std_obj":["Arabic","Math","History","Geography","Sicence"],
           "std_name":["Ahmed","Eyad","Khalid","Ishaq","Saleh"],
           "std_laat":["Ishaq","Jamal","Mohamed","Zayed","Ali"]})
studentdf = pd.DataFrame(students_info)
studentdf
```

```python
# using applymap with len
studentdf.applymap(len)
# the code above will give us the "len" of every invisiual row within our dataframe
```

SO THE CONCLUSION IS THAT WE BETTER USE "applymap()" WHEN WE WANT TO PERFORM SPECIFIC TASK OR FUNCTION ON THE HOLE DATFRAME AT ONCE, we just neeed to make sure that all the dataframe values are same data type

#####################################################################################################

(2)- Grouping our data:

A groupby operation involves some combination of: 1- splitting the object, 2-applying a function, 3-combining the results, This can be used to group large amounts of data and performe operations on these groups.

```python
# 1- splitting the object: here we choose the columns we want to group on,exa: "Position" column

Position_grp = df1.groupby(["Position"])
# code above will return a groupby object that contains a bunch of groups based on the "Position" colunm
Position_grp
```

```python
#  now lets taik a look on one of these group in the created dataframe object:

Position_grp.get_group("Pitcher")
# code above will print "Pitcher" group, now since we grouped based on "Position" column, each 
#... indvisual unique position will be a seperate group
```

```python
# 2-applying a function : now we can apply any fuction we want on these groups we created,exa:

Position_grp.get_group("Pitcher").mean().round()
# code above will print the "maen" salary for the "Pitcher" position
```

```python
# also we can apply a function in a specific column in the groupby dataframe, exa:

Position_grp["Salary"].mean().round()
# code above will print the "mean" salary for each indvisual unique position
```

```python
# another example: 

Position_grp["Team"].value_counts().head(60)
# code above will print how many player every team has in every position 
```

```python
# we can use the "agg" to apply a multiple aggregation functions at the same time,exa:

Position_grp["Salary"].agg(["median","mean"]).round()
# code above will print the "median" and "mean" for the salary
```
